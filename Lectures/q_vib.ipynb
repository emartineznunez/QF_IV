```python
%matplotlib inline
from matplotlib import pyplot
import numpy as np

#S. I. units freq of 1000 cm-1
h = 6.626e-34
c = 3e8
k_B = 1.38e-23
nu = 100000
 
T = np.linspace(0.1,1000,num=100)
y = h * c * nu / (k_B * T)
q_v = 1 / (1 - np.e**(-y) )

pyplot.plot(T,q_v,'-',label="vibrational partition function")
pyplot.legend()
pyplot.xlabel("T (K)")
pyplot.ylabel("q_vib")
```
