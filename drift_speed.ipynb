```python
%matplotlib inline
from matplotlib import pyplot
import numpy as np

Z = 2
e = 1.6e-19
E = 1000 # volts / m
m = 24.3 / 6.022e23 # g
eta = 1e-2 # g cm^-1 s
r = 1e-8 # cm

A = Z * e * E / m
B = 6 * np.pi * eta * r / m
t = np.linspace(0,1e-12,num=100)
v = A / B * (1 - np.e**(-B * t) ) * 100 # cm
pyplot.plot(t,v,'-',label="drift speed")
pyplot.legend()
pyplot.xlabel("time (s)")
pyplot.ylabel("v(cm/s)")
print('Limiting speed:',A / B * 100,'cm/s')
#To reach 99% of the limiting speed
p = float(input('Percentage: '))
x = 100 / (100 - p)
print('Time to reach 99% of limiting speed:', np.log(x) / B,'s')
```
